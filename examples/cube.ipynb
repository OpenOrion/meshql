```python
import cadquery as cq
from meshql import GeometryQL

with GeometryQL() as geo:
    (
        geo
        .load(
            (
                cq.Workplane("XY")
                .box(10,10,10)
                .rect(2, 2)
                .cutThruAll()
            ),
            on_split = lambda split: (
                split
                .from_plane(angle=(90,90,0))
                .from_plane(angle=(-90,90,0))
            ),
            use_cache=True
        )
        .setTransfiniteAuto(max_nodes=300)
        .generate(3)
        .show("gmsh")
    )
```

# With Boundary Layer

```python
import cadquery as cq
from meshql import GeometryQL


with GeometryQL() as geo:
    (
        geo
        .load(
            (
                cq.Workplane("XY")
                .box(10,10,10)
                .rect(2, 2)
                .cutThruAll()
            ),

            on_split = lambda split: (
                split
                .from_plane(angle=(-90,90,0))
                .from_plane(angle=(90,90,0))
            )
        )
        .setTransfiniteAuto(300)

        .faces(type="interior")
        .addBoundaryLayer(0.001)
        .end()

        .generate(3)
        .show("gmsh")
    )

```
